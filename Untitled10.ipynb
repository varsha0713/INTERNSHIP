```python
import pandas as pd
import scipy.stats as stats

# Assuming you've already loaded your data and split it into X (independent variables) and y (dependent variable)

# Create DataFrames for the independent and dependent variables
independent_variables = pd.DataFrame(X, columns=X.columns)
dependent_variable = pd.DataFrame(y, columns=['label'])

# Create a contingency table
contingency_table = pd.crosstab(independent_variables['service'], dependent_variable['label'])

# Perform the chi-square test of independence
chi2, p, _, _ = stats.chi2_contingency(contingency_table)

# Print the results
print("Chi-squared statistic:", chi2)
print("P-value:", p)

# Interpret the results based on the p-value and significance level (e.g., 0.05)
if p < 0.05:
    print("There is a significant association between the independent and dependent variables.")
else:
    print("There is no significant association between the independent and dependent variables.")
```
